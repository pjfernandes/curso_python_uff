```python
frutas = ["maçã", "abacaxi", "laranja"]

# Adicionar elemento no final da lista
frutas.append("mamão")

print(frutas)
```

```python
# Remover elemento da lista pelo nome
frutas.remove("abacaxi")
print(frutas)
```

```python
# Ordenar a lista
frutas.sort()
print(frutas)
```

# **3) Tuplas**
São parecidas com a listas, porém são imutáveis e são declaradas utilizando **()**

https://www.geeksforgeeks.org/python-tuples/


```python
numeros = (1, 6, 4, 5, 2, 4, 3, 1, 4, -1)

# Obter o comprimento da tupla
tamanho = len(numeros)

# Contar ocorrências de um elemento
ocorrencias = numeros.count(4)

print(f"Tamanho da Tupla: {tamanho}")
print(f"Ocorrências de '4': {ocorrencias}")
```

# **Exercícios**

1. Escreva um programa que gere como resultado um dicionário com o número de ocorrência de palavras no texto de uma receita via API. Imprima o dicionário.

**Não esqueça de remover a pontuação e colocar em letras minúsculas**

```python
#Para remover a pontuação
import string
s = "às vezes! nome. pedro. João"
sem_pontuacao = s.translate(str.maketrans('', '', string.punctuation))
print(sem_pontuacao)
```

```python
import requests
import json

url_api = "https://www.themealdb.com/api/json/v1/1/random.php"
response = requests.get(url_api)
dados = json.loads(response.text)
```

```python
receita = dados['meals'][0]['strInstructions']
```

```python
import string
palavras = receita.lower().translate(str.maketrans('', '', string.punctuation))
palavras = palavras.split()
print(palavras)
```

```python
d = {}

for palavra in palavras:
  if palavra not in d:
    d[palavra] = 1
  else:
    d[palavra] += 1
```

```python
for key, value in d.items():
  print(f"{key}: {value}")
```
