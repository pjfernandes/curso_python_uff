**Função para encontrar valor máximo**

```python
def maximo(lista):
  maximo = None
  for numero in lista:
    if maximo == None or maximo < numero:
      maximo = numero
  return maximo


```

```python
maximo([-11, -12, -2])
```

**Função para encontrar valor**

```python
def encontrar(valor, lista):
  for numero in lista:
    print(numero)
    if numero == valor:
      return True

  return False
```

**Fatorial não recursivo**

<img src="https://s3.static.brasilescola.uol.com.br/be/2020/09/multiplicacao-fatorial.jpg"/>

Fonte: https://brasilescola.uol.com.br/matematica/fatorial.htm

```python
def fatorial(n):
  resultado = 1
  for numero in list(range(1, n + 1)):
    resultado = resultado * numero

  return resultado
```

```python
fatorial(5)
```

**Fatorial recursivo**

Um procedimento é recursivo quando contém “uma ou mais chamadas a si mesmo” (SZARCFITER; MARKENZON, 2011)

<img src="https://media.geeksforgeeks.org/wp-content/uploads/20191108122114/Untitled-Diagram58.jpg"/>

Fonte: https://www.geeksforgeeks.org/introduction-to-recursion-data-structure-and-algorithm-tutorials/

```python
def fatorial_recursivo(n):
  if n == 0 or n == 1:
    return 1
  else:
    return n * fatorial(n - 1)
```

```python
fatorial_recursivo(5)
```
