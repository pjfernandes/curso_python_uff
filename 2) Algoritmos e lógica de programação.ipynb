# **2) Expressões aritméticas**

```python
a = 5
b = 3

soma = a + b
subtracao = a - b
multiplicacao = a * b
divisao = a / b
resto_divisao = a % b

print(f"Soma: {soma}")
print(f"Subtração: {subtracao}")
print(f"Multiplicação: {multiplicacao}")
print(f"Divisão: {divisao}")
print(f"Resto da Divisão: {resto_divisao}")
```

# **3) Lógica booleana:**


```python
verdadeiro = True
falso = False

and_operador = verdadeiro and falso
or_operador = verdadeiro or falso
not_operador = not verdadeiro

print(f"AND: {and_operador}")
print(f"OR: {or_operador}")
print(f"NOT: {not_operador}")
```

# **4) Expressões relacionais**

```python
x = 10
y = 5

igual = x == y
diferente = x != y
maior_que = x > y
menor_que = x < y

print(f"Igual: {igual}")
print(f"Diferente: {diferente}")
print(f"Maior que: {maior_que}")
print(f"Menor que: {menor_que}")
```

# **8.1) While loop**

 Executa a repetição enquanto uma condição for verdadeira

```python
contador = 1

while contador <= 5:
    print(f"Contagem: {contador}")
    #contador = contador + 1
    contador += 1
```

# **9) Implementação de algoritmo para encontrar o valor máximo em uma lista**

O que fazer?

1. Criar uma variável que irá armazenar o valor máximo
2. Percorrer cada item da lista
3. Se esse item for maior que o que está armazenado na variável com o valor máximo, coloque o novo valor

```python
numeros = [3, 4, 2, 0, 10, 22, 34, 15]
```

```python
max = 0

for numero in numeros:
  if numero > max:
    max = numero
```

```python
print(max)
```

# **10) Simulação de jogo de dados. Jogar o dado até que se obtenha o valor 5**

<img src="https://upload.wikimedia.org/wikipedia/commons/a/a5/6sided_dice.jpg"/>

```python
faces_do_dado = [1, 2, 3, 4, 5, 6]
```

```python
import random
```

```python
random.choice(faces_do_dado)
```

```python
jogada = random.choice(faces_do_dado)

while jogada != 5:
  print(jogada)
  jogada = random.choice(faces_do_dado)
```
