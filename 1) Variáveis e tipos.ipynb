# **7.1) Programa para converter temperatura Celsius para Kelvin**

```python
temperatura_celsius = float(input("Digite a temperatura em graus Celsius: "))
temperatura_kelvin = temperatura_celsius + 273.15

print(f"A temperatura {temperatura_celsius} graus Celsius é igual a {temperatura_kelvin} Kelvin.")
```
